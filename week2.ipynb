```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

```python
#4. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!
imageWithRectangle = cv.cvtColor(img,cv.COLOR_BGR2RGB)
fig = plt.figure(figsize=(10,10))
cv.rectangle(imageWithRectangle, pt1=(0,75), pt2=(75,0), color=(0,255, 255), thickness = 5)
cv.rectangle(imageWithRectangle, pt1=(76,150), pt2=(150,76), color=(0,255,0), thickness = 5)
cv.rectangle(imageWithRectangle, pt1=(300,200), pt2=(260,250), color=(255,255,255), thickness = 5)
cv.rectangle(imageWithRectangle, pt1=(200,250), pt2=(250,200), color=(255,0,0), thickness = 5)
cv.rectangle(imageWithRectangle, pt1=(150,200), pt2=(200,150), color=(0,0,255),thickness=5)
plt.imshow(imageWithRectangle)
```

```python
#5. Tampilkan image dengan posisi terbalik!
img_t = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.imshow(img_t)
```

```python
#6. Buat rectangle dan circle pada bagian wajah dari image kucingku!
img = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/kitten01.jpg')
fig = plt.figure(figsize=(5,5))

ax = fig.add_subplot(111)
cv.rectangle(img, pt1=(250,30), pt2=(100,140), color=(255, 0, 30), thickness = 5)
cv.circle(img, center=(175, 75), radius=80, color=(180, 105, 255), thickness = 6)
ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```
